#### Load the pickel file for level 2 - Feature extraction

```python
import numpy as np
import pandas as pd
import pickle
```

```python
matches = pickle.load(open('dataset_level1.pkl','rb'))
matches.iloc[0]['innings'][0]['1st innings']['deliveries']   # data of single innings
```

### Creat Clean DataFrame using the datas(expected columns in the output dataframe:
#### batting_team, bowling team, city, current score, ball left, wickets left, current run rate, last five overs total runs

```python
count = 1
df3 = pd.DataFrame()
for index, row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count+=1
        continue
    count+=1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    loop_df = pd.DataFrame({
            'match_id':match_id,
            'teams':teams,
            'batting_team':batting_team,
            'ball':ball_of_match,
            'batsman':batsman,
            'bowler':bowler,
            'runs':runs,
            'player_dismissed':player_of_dismissed,
            'city':city,
            'venue':venue
        })
    df3 = df3.append(loop_df)
```

```python
df3
```

#### Extracting bowling team from teams column

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
df3['bowling_team'] = df3.apply(bowl,axis=1)
```

```python
df3
```

```python
df3.drop(columns=['teams'],inplace=True)
```

```python
df4  = df3.copy()
```

```python
df4['batting_team'].unique()
```

```python
teams = [
    'Sri Lanka',
    'Australia',
    'India',
    'Bangladesh',
    'New Zealand',
    'South Africa',
    'England',
    'West Indies',
    'Afghanistan',
    'Pakistan'
]
```

```python
df6
```

#### Handling NAN Values

```python
df6.isnull().sum()
```

```python
df6[df6['city'].isnull()]['venue'].value_counts()
```

```python
cities = np.where(df6['city'].isnull(),df6['venue'].str.split().apply(lambda x:x[0]),df6['city'])
```

```python
df7 = df6.copy()
```

```python
df7['city'] = cities
```

```python
df7.isnull().sum()
```

```python
df7.drop(columns=['venue'],inplace=True)
df7
```

#### Only considering the Cities where played more than 600 balls 

```python
df7['city'].value_counts()
```

```python
eligible_cities = df7['city'].value_counts()[df7['city'].value_counts() > 600].index.tolist()
```

```python
df8 = df7[df7['city'].isin(eligible_cities)]
df8
```

#### Extracting current_score column from runs

```python
df8['current_score'] = df8.groupby('match_id').cumsum()['runs']
df8
```

```python
df9 = df8.copy()
```

#### Extracting balls_left column from the data

```python
df9['over'] = df9['ball'].apply(lambda x:str(x).split(".")[0])
df9['ball_no'] = df9['ball'].apply(lambda x:str(x).split(".")[1])
df9
```

```python
df9['balls_bowled'] = (df9['over'].astype('int')*6) + df9['ball_no'].astype('int')
df9
```

```python
df9['balls_left'] = 120 - df9['balls_bowled']
df9['balls_left'] = df9['balls_left'].apply(lambda x:0 if x<0 else x)
df9
```

```python
df10 = df9.copy()
```

#### Extracting wickets_left column from the data

```python
df10['player_dismissed'] = df10['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df10['player_dismissed'] = df10['player_dismissed'].astype('int')
df10['player_dismissed'] = df10.groupby('match_id').cumsum()['player_dismissed']
df10['wickets_left'] = 10 - df10['player_dismissed']
df10
```

```python
df11 = df10.copy()
```

#### Extracting crr column from the data

```python
df11['crr'] = (df11['current_score']*6)/df11['balls_bowled']
df11
```

#### Extracting last_five column from the data

```python
groups = df11.groupby('match_id')

match_ids = df11['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
df11['last_five'] = last_five
df11
```

#### Extracing run_x column(final runs) from the data

```python
df12 = df11.groupby('match_id').sum()['runs'].reset_index().merge(df11,on='match_id')
df12
```

```python
df13=df12[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

#### droping rows that contain the NAN values in last_five column

```python
df13.dropna(inplace=True)
```
